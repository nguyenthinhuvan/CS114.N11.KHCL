# **Extract data from Tuoitrecuoi HTML source**

## **Find all article section with link to News page from cover HTML file**

### **Read HTML file**

```python
with open('/content/gdrive/MyDrive/Máy học/News Headlines Sarcasm Detection/tuoitrecuoiHTMLsource.txt', 'r') as f:
  coverHTML = f.read()
```

### **Split article News**

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(coverHTML, 'html5lib')
```

```python
cover_article = soup.find_all('article', class_="art-4 mb-4 news-item")
```

### **Process get Date of News + headlines + URL**

**Setting Webdrive**

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

**Main Process**

```python
len(cover_article)
```

```python
import time
```

```python
headline_News = []
NewsWebsite_url = 'https://cuoi.tuoitre.vn'
Times = 0

for i in cover_article:
  Times += 1
  # if Times == 100:
  #   driver.close()
  #   chrome_options = webdriver.ChromeOptions()
  #   chrome_options.add_argument('--headless')
  #   chrome_options.add_argument('--no-sandbox')
  #   driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
  #   Times = 0

  data_ourl = i['data-ourl']
  article_url = NewsWebsite_url + data_ourl
  driver.get(article_url)
  XPathLoading = "//div[@class='top-art']"
  announce = WebDriverWait(driver, 30).until(EC.presence_of_element_located((By.XPATH, XPathLoading)))

  coverHTML_News = driver.page_source
  soup = BeautifulSoup(coverHTML_News, 'html5lib')
  HTML_data = soup.find('div', class_='top-art')
  
  headline = HTML_data.h1.contents
  date_upload = HTML_data.div.p.contents

  print(Times, date_upload, headline, article_url)
  headline_News.append([Times - 1, date_upload, headline, article_url])
```

**End Process and close browser**

```python
driver.close()
```

## **Save data as json file**

```python
headline_News
```

```python
import json

f = open('/content/gdrive/MyDrive/Máy học/News Headlines Sarcasm Detection/tuoitrecuoiNewsHeadlinesData.json', 'w')
for i in headline_News:
  id = i[0]
  date = i[1][0][9:]
  headline = i[2][0]
  url = i[3]
  languageCode = 'vie | ISO 639-2 Code'
  element = {
      'id': id,
      'headline': i[2][0],
      'article url': i[3],
      'is_sarcastic': 1,
      'date': i[1][0][9:],
      'lang': 'vie | ISO 639-2 Code'
  }

  json.dump(element, f)
  f.write("\n")
```
